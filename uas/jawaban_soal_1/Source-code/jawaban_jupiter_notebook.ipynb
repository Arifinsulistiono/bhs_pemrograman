## CODE REKURSIF

```python
def hitung_fibonacci(n):
    if n <= 1:
        return n
    else:
        return hitung_fibonacci(n-1) + hitung_fibonacci(n-2)

# Menampilkan Fibonacci ke-7
print(hitung_fibonacci(7))
```
